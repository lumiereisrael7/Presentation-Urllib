Nous pouvons le voir dans un exemple :</p>

 <div style="text-align: justify; font-family: Georgia, serif; font-size: 20px; color: purple">urllib.parse : </div>
 <p style="text-align: justify; font-family: Georgia, serif; font-size: 17px">Ce module définit des fonctions pour manipuler les URL et leurs composants, pour les construire ou les casser. Il se concentre généralement sur la division d'une URL en petits composants ; ou joindre différents composants d'URL dans une chaîne d'URL.
Nous pouvons le voir dans le code ci-dessous :
Nous pouvons le voir dans un exemple :</p>

```python
from urllib.parse import * 

parse_url = urlparse('https://github.com/lumiereisrael7/Presentation-urllib/blob/9d818926269592754e808a304b89c11e2c4d6485/README.md')
print(parse_url)
print("\n")
unparse_url = urlunparse(parse_url)
print(unparse_url)
```

 <div style="text-align: justify; font-family: Georgia, serif; font-size: 20px; color: purple">urllib.error : </div>
 <p style="text-align: justify; font-family: Georgia, serif; font-size: 17px"> Ce module définit les classes d'exception levées par urllib.request. Chaque fois qu'il y a une erreur dans la récupération d'une URL, ce module aide à lever des exceptions. Voici les exceptions soulevées :
</p>

### URLError 
    - Il est déclenché pour les erreurs dans les URL, ou les erreurs lors de la récupération de l'URL en raison de la connectivité, et a une propriété « raison » qui indique à un utilisateur la raison de l'erreur.

### HTTPError
    – Il est déclenché pour les erreurs HTTP exotiques, telles que les erreurs de demande d'authentification. C'est une sous-classe ou URLError. Les erreurs typiques incluent « 404 » (page non trouvée), « 403 » (demande interdite) et « 401 » (authentification requise).

```python
# URL Error

import urllib.request
import urllib.parse

# trying to read the URL but with no internet connectivity
try:
    x = urllib.request.urlopen('https://www.google.com')
    print(x.read())

# Catching the exception generated
except Exception as e :
    print(str(e))
```

 <div style="text-align: justify; font-family: Georgia, serif; font-size: 20px; color: purple">urllib.robotparser: </div>
 <p style="text-align: justify; font-family: Georgia, serif; font-size: 17px">Ce module contient une seule classe, RobotFileParser. Cette classe répond à la question de savoir si un utilisateur particulier peut ou non récupérer une URL qui a publié des fichiers robot.txt. Robots.txt est un fichier texte que les webmasters créent pour indiquer aux robots Web comment explorer les pages de leur site Web. Le fichier robot.txt indique au grattoir Web quelles parties du serveur ne doivent pas être accédées.
Par example :</p>

```python
# importing robot parser class
import urllib.robotparser as rb

bot = rb.RobotFileParser()

# checks where the website's robot.txt file reside
x = bot.set_url('https://www.geeksforgeeks.org/robot.txt')
print(x)

# reads the files
y = bot.read()
print(y)

# we can crawl the main site
z = bot.can_fetch('*', 'https://www.geeksforgeeks.org/')
print(z)

# but can not crawl the disallowed url
w = bot.can_fetch('*', 'https://www.geeksforgeeks.org/wp-admin/')
print(w)
```
